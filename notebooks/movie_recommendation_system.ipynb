```python
# required columns
# genres, keywords, overview, title, movie_id, cast, crew, vote_average
```

```python
df = df[['movie_id', 'title', 'genres', 'overview', 'keywords', 'cast', 'crew', 'vote_average']]
```

```python
df.dropna(inplace = True)
```

```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(1)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
df.head()
```

```python
import ast

def cast_convert(text):
    s = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            s.append(i['name'])
        counter += 1
    return s
```

```python
df['cast'] = df['cast'].apply(cast_convert)
```

```python
df.head()
```

```python
df['crew'][0]
```

```python
import ast

def crew_convert(text):
    a = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            a.append(i['name'])
    return a
```

```python
df['crew'] = df['crew'].apply(crew_convert)
```

```python
df.head()
```

```python
def collapse(text):
    l = []
    for i in text:
        l.append(i.replace(' ', ''))
    return l
```

```python
df['genres'] = df['genres'].apply(collapse)
df['keywords'] = df['keywords'].apply(collapse)
df['cast'] = df['cast'].apply(collapse)
df['crew'] = df['crew'].apply(collapse)
```

```python
df.head()
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
df = df.drop(columns = ['overview', 'keywords', 'cast', 'crew'], axis = 1)
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x : " ".join(x))
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vector = cv.fit_transform(df['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```
